# Salary

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv("salary_data.csv")
df.head()
```

# Trained The Model

```python
x = df[['Experience']]
y = df[['Salary']]
```

```python
X_train,x_test,Y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2=r2_score(y_test,y_pred)
print(mse)
print("Accuracy:"), r2
```

```python
plt.scatter(x_test, y_test, color ='blue', label = "Actual Value")
plt.plot(x_test, y_pred, color ='black', label = "Actual Value")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience VS Salary")
plt.legend()
plt.show()
```

# Predict Salary


```python
def predict_salary(exp):
    predict_sal = model.predict(np.array([[exp]]))
    return predict_sal[0][0]
    
experience = float(input("Enter the experience:"))
predicted_salary = predict_salary(experience)
print("Predict Salary:", predicted_salary)
```

```python
df = pd.read_csv("Housing_Price.csv")
df.head()
```

```python
a = df[['bedrooms']]
b = df[['price']]
```

```python
A_train,a_test,B_train,b_test=train_test_split(a,b,test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(A_train, B_train)
```

```python
b_pred = model.predict(a_test)
```

```python
mse = mean_squared_error(b_test, b_pred)
r2=r2_score(b_test,b_pred)
print(mse)
print("Accuracy:"), r2
```

```python
plt.scatter(a_test, b_test, color ='blue', label = "Actual Value")
plt.plot(a_test, b_pred, color ='black', label = "Actual Value")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms VS Price")
plt.legend()
plt.show()
```

```python
def predict_housing(exp):
    predict_housing = model.predict(np.array([[exp]]))
    return predict_housing[0][0]
    
experience = float(input("Enter the experience:"))
predicted_housing = predict_housing(experience)
print("Predict Housing:", predicted_housing)
```
